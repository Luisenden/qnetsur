```python
import networkx as nx
import matplotlib.pyplot as plt

def create_random_tree(n):
    # Create a random tree
    graph = nx.random_tree(n, seed=42)
    
    # Draw the tree
    pos = nx.spring_layout(graph, scale=2)
    nx.draw(graph, pos, with_labels=False, node_color='lightblue', node_size=50)
    plt.show()

# Number of nodes in the tree
n = 100
create_random_tree(n)
```

```python
n=100
graph = nx.random_tree(n, seed=7)

fig, ax = plt.subplots(figsize=(15,7))
pos = nx.spring_layout(graph, scale=2)
nx.draw(graph, pos, with_labels=True, node_color='lightblue', node_size=150)
plt.tight_layout()
```
