```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
%matplotlib inline
plt.style.use("seaborn-v0_8-paper")

font = 16
plt.rcParams.update({
    'text.usetex': False,
    'font.family': 'arial',
    'font.size': font,
    'axes.labelsize': font,  
    'xtick.labelsize': font,  
    'ytick.labelsize': font, 
    'legend.fontsize': font,
    'legend.title_fontsize': font,
    'axes.titlesize': font
})
import warnings
warnings.filterwarnings("ignore")
```

```python
path_to_output_data_folder = ''
```

```python

path = path_to_output_data_folder+'/continuous_distribution_protocols/notebook_cd_n3/SU_tree-2-1_100cycles_SEED42_06-26-2024_16:30:08.csv'
#path = path_to_output_data_folder+'/continuous_distribution_protocols/notebook_cd_n10/SU_randtree-10_100cycles_SEED42_08-26-2024_11:53:41.csv'
data = pd.read_csv(path)
```

```python
sns.lineplot(data=data, x='Iteration', y='DecisionTree', errorbar='ci', marker='o')
sns.lineplot(data=data, x='Iteration', y='SVR', errorbar='ci', marker='^')
plt.ylabel('MAE')
plt.grid()
plt.yscale('log')
plt.ylim(0.01, 1)
plt.xlabel(r'$t$')
plt.savefig('cd_n3.pdf', dpi=300, bbox_inches='tight')
```

```python
files = glob.glob(path_to_output_data_folder+'/continuous_distribution_protocols/cd_n100/cd_10h/SU_*')

data = []
for i, file in enumerate(files):
    df = pd.read_csv(file)
    df['File'] = i
    data.append(df)
data = pd.concat(data)
```

```python
sns.lineplot(data=data, x='Iteration', y='DecisionTree', errorbar='ci', label='Random Forest Regression', marker='o')
sns.lineplot(data=data, x='Iteration', y='SVR', errorbar='ci', label='Support Vector Regression', marker='^')
plt.ylabel('')
plt.grid()
plt.yscale('log')
plt.ylim(0.099, 1)
plt.xlabel(r'$t$')
plt.savefig('cd_n100.pdf', dpi=300, bbox_inches='tight')
```
