```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-v0_8')
import numpy as np
```

Exploration / Exploitation

```python
niter = 50
f = lambda count: (1-np.log(1+count/int(niter))**2)**4
plt.plot([f(count) for count in np.linspace(0,int(niter))])
```

```python
from scipy.stats import truncnorm

par = (2,7)
x = 4.3
std = f(10) * (par[1] - par[0])/2
sample = truncnorm.rvs((par[0] - x) / std, (par[1] - x) / std, loc=x, scale=std, size=10000)
sample = sample.astype(int)
```

```python
fig = plt.hist(sample, bins=50)
plt.xlim((0,9))
plt.show()
```
