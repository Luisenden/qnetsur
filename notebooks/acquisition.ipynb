```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn-v0_8')
import numpy as np
```

### Exploration / Exploitation

```python
T = 1

f = lambda count, k: (1-np.log(1+count/int(T))**2)**k
x = np.linspace(0,T,100)
for k in range(0,6):
    plt.plot(x, [f(count,k) for count in x], label=f'k={k}', color=sns.color_palette()[k])
plt.legend()
```

```python
from scipy.stats import truncnorm

par = (2,7)
x = 4.3
std = f(10) * (par[1] - par[0])/2
sample = truncnorm.rvs((par[0] - x) / std, (par[1] - x) / std, loc=x, scale=std, size=10000)
sample = sample.astype(int)
```

```python
fig = plt.hist(sample, bins=50)
plt.xlim((0,9))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
p = 0.9
t = np.linspace(0,3, 100)
m = 0.2
styl = [':', '--', '-']
for i, m in enumerate([0.1,0.2,0.5]):
    plt.plot(t, np.exp(p-p*t/m), label=f'upper m={m}', linestyle = styl[i], color='black')
    plt.plot(t, np.exp(-p*t/m*1/(1-p)), label=f'lower m={m}', linestyle = styl[i], color='black')
plt.ylim(0,1)
plt.ylabel(r'$P(T_{output}>t)$')
plt.xlabel(r't')
plt.title(r'Bounds with $p=0.9$')
plt.tight_layout()
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
p = 0.1
t = np.linspace(0,3, 100)
m = 0.2
styl = [':', '--', '-']
for i, m in enumerate([0.1,0.2,0.5]):
    plt.plot(t, np.exp(p-p*t/m), label=f'upper m={m}', linestyle = styl[i],  color='black')
    plt.plot(t, np.exp(-p*t/m*1/(1-p)), label=f'lower m={m}', linestyle = styl[i], color='black')
plt.ylim(0,1)
plt.ylabel(r'$P(T_{output}>t)$')
plt.xlabel(r't')
plt.title(r'Bounds with $p=0.1$')
plt.tight_layout()
```

```python
font = 22
plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
    'font.size': font,
    'axes.labelsize': font,  
    'xtick.labelsize': font,  
    'ytick.labelsize': font, 
    'legend.fontsize': font,
    'legend.title_fontsize': font,
    'axes.titlesize': font
})
t = np.linspace(1,10,100)
plt.plot(t, (1-p)**np.floor(t), label=r'$T_{gen}$')
plt.plot(t, np.exp((t-1)*np.log(1-p)), label=r'$T^{upper}_{gen}$')
plt.xlabel('t')
plt.ylabel(r'$Pr(T>t)$')
plt.legend()
```
